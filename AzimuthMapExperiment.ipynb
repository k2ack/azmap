Let's load my log as a simple CSV -- would change this to read ADIF once I find an ADIF parser that isn't AGPL.

* *contact_grids* is a filtered DataFrame with only items that had a gridsquare.
* I split into two groups since I want to colorize.
* Collect all DXCCs  into contact_dxcc
* The set difference is in plot_by_dxcc and these are the only things I need to plot by DXCC location



```python
load_path = 'locations_test.csv'

contacts = pd.read_csv(load_path,header=0)
contact_with_grids = contacts[ contacts.Grid.notnull() ]
contact_locs = contact_with_grids.Grid
contact_dxcc = set(contacts.DXCC)
plotted_by_grid = set(contact_with_grids.DXCC)
plot_by_dxcc=contact_dxcc.difference(plotted_by_grid)
dxcc_grids = [ loc_for_dxcc(d) for d in plot_by_dxcc ]

print ('total dxcc: {}, grids for: {}, remain by dxcc: {}'.format(len(contact_dxcc),len(plotted_by_grid),len(plot_by_dxcc)))
#print plot_by_dxcc
#print dxcc_grids
```

Now we can draw a map in Azimuth Equidistant Projection centered on our QTH and plot contacts by grid, then by dxcc.

```python
def azmap(qth):
    proj='aeqd'
    lat,lon = mlocs.toLoc(my_qth)
    m = Basemap(projection=proj,lat_0=lat,lon_0=lon)
    m.drawmapboundary()
    m.drawcountries(linewidth=0.25)
    m.drawcoastlines(linewidth=0.5)
    m.drawparallels(np.arange(-80.0,81,20),linewidth=0.25)
    m.drawmeridians(np.arange(-180,180,20),linewidth=0.25)
    #plt.title('Contact Locations by Grid Square from {}'.format(my_qth))
    return m

m = azmap(my_qth)
# Draw GCR lines to each contact
draw_circle(m, my_qth, contact_locs)
draw_circle(m, my_qth, dxcc_grids)

# Draw our markers for contacts.
draw_qth( m, contact_locs,   color='blue')
draw_qth( m, all_qths, color='red', marker='8')
draw_qth( m, dxcc_grids, alpha=0.5, marker='h')

```
